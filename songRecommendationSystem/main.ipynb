```python
import numpy as np
import pandas as pd
import ast
```

```python
df = pd.read_csv('ex.csv',encoding='cp1252')
df.head()
#df=df.head(2000)
```

```python
df.head()
```

```python
def handel_genre(obj):
    obj = str(obj)
    lst = list(obj.split(','))
    res = []
    for i in lst:
        i = str(i)
        i=i.replace(" ",'')
        res.append(i)
    return res
```

```python
df['Singer/Artists']=df['Singer/Artists'].apply(handel_genre)
df['Album/Movie']=df['Album/Movie'].apply(lambda x:[str(x).replace(" ","")])
df['Genre']=df['Genre'].apply(lambda x:[str(x).replace(" ","")])
```

```python
df.head()
```

```python
df['tags']=df['Singer/Artists']+df['Genre']+df["Album/Movie"]
```

```python
df = df.drop(['Genre','User-Rating','Singer/Artists','Album/Movie'],axis=1)
df.head()
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2000)#no need of stopword parameter bcz movie names are in hindi 
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

Uding cosine distance

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend_song(song):
    song_index = df[df['Song-Name']==song].index[0]
    distance = similarity[song_index]
    recommended_songs = sorted(list(enumerate(distance)),reverse=True,key = lambda x:x[1])[1:6]
    #print(recommended_songs)
    for i in recommended_songs:
        print(df['Song-Name'][i[0]])
```

```python
recommend_song('Bom Diggy Diggy')
```
